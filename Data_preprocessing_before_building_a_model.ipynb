```python
# Missing value treatment
#choose the method of imputing missing values

df.isnull().sum()
```

```python
for i in ['Age','Cabin']:
    df[i].fillna(df[i].mode(),inplace=True)
```

```python
from sklearn.impute import KNNImputer
impute = KNNImputer()
```

```python
for i in df.select_dtypes(include='number').columns:
    df[i]=impute.fit_transform(df[[i]])
```

```python
# Outliers treatments
# decide whether to do outliers treatment or not, if do how?

def wisker(col):
  q1,q3=col.quantile([0.25,0.75])
  iqr=q3-q1
  lower_bound=q1-(1.5*iqr)
  upper_bound=q3+(1.5*iqr)
  return lower_bound,upper_bound
```

```python
wisker(df['Age'])
```

```python
for i in ['Age']:  # Apply outlier treatment only to 'Age'
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]>uw,uw,np.where(df[i]<lw,lw,df[i]))
```

```python
df.columns
```

```python
# Duplicates & garbage value treatment
# check for duplicate if we have any unique column in the date set, delete them
# clean the garbage value

df.drop_duplicates(inplace=True)
```

```python
# Encoding of data
# do label encoding and one hot encoding with pd.getdummies
df=pd.get_dummies(df,columns=['Sex','Embarked'],drop_first=True)
```
